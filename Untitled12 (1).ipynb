```python
import json
import cv2
import numpy as np
from pathlib import Path
from tqdm import tqdm

# =========================
# PATHS (COLAB VERSION ✅)
# =========================
KAGGLE_ROOT = Path("/content/TUSimple/train_set")

OUT_ROOT = Path("/content/tusimple_processed")
OUT_IMG_DIR = OUT_ROOT / "images"
OUT_MSK_DIR = OUT_ROOT / "masks"

OUT_IMG_DIR.mkdir(parents=True, exist_ok=True)
OUT_MSK_DIR.mkdir(parents=True, exist_ok=True)

LABEL_FILES = [
    KAGGLE_ROOT / "label_data_0313.json",
    KAGGLE_ROOT / "label_data_0531.json",
    KAGGLE_ROOT / "label_data_0601.json"
]

# =========================
# SETTINGS
# =========================
LANE_THICKNESS = 20

total_saved = 0
missing = 0

print("Starting preprocessing...")

# =========================
# MAIN LOOP
# =========================
for label_file in LABEL_FILES:

    if not label_file.exists():
        print(f"Skipping missing label file: {label_file}")
        continue

    print(f"\nProcessing: {label_file}")

    with open(label_file) as f:
        lines = f.readlines()

    for line in tqdm(lines):

        data = json.loads(line)

        # ✅ Correct path (already relative to train_set)
        img_path = KAGGLE_ROOT / data["raw_file"]

        if not img_path.exists():
            missing += 1
            continue

        img = cv2.imread(str(img_path))
        if img is None:
            missing += 1
            continue

        h, w, _ = img.shape
        mask = np.zeros((h, w), dtype=np.uint8)

        # =========================
        # DRAW LANES
        # =========================
        for lane in data["lanes"]:

            pts = [
                (x, y)
                for x, y in zip(lane, data["h_samples"])
                if x >= 0
            ]

            if len(pts) > 1:
                pts = np.array(pts, np.int32)

                # smoother continuous lane
                cv2.polylines(
                    mask,
                    [pts],
                    isClosed=False,
                    color=255,
                    thickness=LANE_THICKNESS
                )

        # =========================
        # SAVE
        # =========================
        name = data["raw_file"].replace("/", "_")

        cv2.imwrite(str(OUT_IMG_DIR / name), img)
        cv2.imwrite(str(OUT_MSK_DIR / name), mask)

        total_saved += 1

# =========================
# DONE
# =========================
print("\nPreprocessing complete")
print("Total saved:", total_saved)
print("Missing images:", missing)
```

```python
from pathlib import Path
import json, cv2, numpy as np
from tqdm import tqdm

KAGGLE_ROOT = Path("/content/TUSimple/train_set")
OUT_ROOT = Path("/content/tusimple_processed")

IMG_DIR = OUT_ROOT / "images"
MSK_DIR = OUT_ROOT / "masks"

IMG_DIR.mkdir(parents=True, exist_ok=True)
MSK_DIR.mkdir(parents=True, exist_ok=True)

LABEL_FILES = [
    KAGGLE_ROOT / "label_data_0313.json",
    KAGGLE_ROOT / "label_data_0531.json",
    KAGGLE_ROOT / "label_data_0601.json"
]

LANE_THICKNESS = 20  # 🔥 critical

total_saved = 0

for label_file in LABEL_FILES:
    with open(label_file) as f:
        lines = f.readlines()

    for line in tqdm(lines):
        data = json.loads(line)

        img_path = KAGGLE_ROOT / data["raw_file"]
        img = cv2.imread(str(img_path))
        if img is None:
            continue

        h, w, _ = img.shape
        mask = np.zeros((h, w), dtype=np.uint8)

        for lane in data["lanes"]:
            pts = [(x, y) for x, y in zip(lane, data["h_samples"]) if x >= 0]

            for i in range(len(pts) - 1):
                cv2.line(mask, pts[i], pts[i+1], 255, LANE_THICKNESS)

        name = data["raw_file"].replace("/", "_")

        cv2.imwrite(str(IMG_DIR / name), img)
        cv2.imwrite(str(MSK_DIR / name), mask)

        total_saved += 1

print("Total saved:", total_saved)
```

```python
import torch
from torch.utils.data import Dataset
from PIL import Image
from pathlib import Path
import numpy as np
import albumentations as A
from albumentations.pytorch import ToTensorV2


class TuSimpleDataset(Dataset):
    def __init__(self, root, split_file, img_size=(256, 512), augment=False):
        self.root = Path(root)
        self.img_dir = self.root / "images"
        self.mask_dir = self.root / "masks"
        self.augment = augment

        with open(self.root / split_file, "r") as f:
            self.files = f.read().splitlines()

        # ---- TRANSFORMS ----
        if self.augment:
            self.transform = A.Compose([
                A.Resize(img_size[0], img_size[1]),
                A.HorizontalFlip(p=0.5),
                A.RandomBrightnessContrast(p=0.3),
                A.GaussNoise(p=0.2),
                A.MotionBlur(p=0.2),
                ToTensorV2()
            ])
        else:
            self.transform = A.Compose([
                A.Resize(img_size[0], img_size[1]),
                ToTensorV2()
            ])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        name = self.files[idx]

        # ---- FULL PATHS ----
        img_path = self.img_dir / name
        mask_path = self.mask_dir / name  # assumes SAME filename

        # ---- LOAD ----
        img = np.array(Image.open(img_path).convert("RGB"))

        # ⚠️ handle missing mask safely
        if mask_path.exists():
            mask = np.array(Image.open(mask_path).convert("L"))
        else:
            # fallback: empty mask (prevents crash)
            mask = np.zeros((img.shape[0], img.shape[1]), dtype=np.uint8)

        # ---- AUGMENT ----
        augmented = self.transform(image=img, mask=mask)

        img = augmented["image"]
        mask = augmented["mask"]

        # ---- FIX SHAPE ----
        if mask.ndim == 2:
            mask = mask.unsqueeze(0)  # (1, H, W)

        # ---- NORMALIZE ----
        img = img.float() / 255.0

        # ---- BINARY MASK ----
        mask = (mask > 0).float()

        return img, mask
```

```python
import torch
import segmentation_models_pytorch as smp

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model = smp.DeepLabV3Plus(
    encoder_name="resnet34",
    encoder_weights="imagenet",
    in_channels=3,
    classes=1,
    activation=None
).to(device)

# better losses for lane segmentation
dice_loss = smp.losses.DiceLoss(mode="binary")
focal_loss = smp.losses.FocalLoss(mode="binary")

# slightly higher LR works better with pretrained encoder
optimizer = torch.optim.Adam(model.parameters(), lr=3e-4)

# learning rate scheduler
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    optimizer,
    mode="min",
    factor=0.5,
    patience=4
)
```

```python
import torch
import torch.nn as nn

# ---- LOSSES ----

bce_loss_fn = nn.BCEWithLogitsLoss()

def dice_loss(preds, targets, smooth=1e-6):
    preds = torch.sigmoid(preds)

    preds = preds.view(-1)
    targets = targets.view(-1)

    intersection = (preds * targets).sum()
    dice = (2. * intersection + smooth) / (preds.sum() + targets.sum() + smooth)

    return 1 - dice


def focal_loss(preds, targets, alpha=0.75, gamma=2):
    preds = torch.sigmoid(preds)

    preds = preds.view(-1)
    targets = targets.view(-1)

    bce = nn.functional.binary_cross_entropy(preds, targets, reduction='none')
    pt = torch.exp(-bce)

    focal = alpha * (1 - pt) ** gamma * bce

    return focal.mean()


# ---- COMBINED LOSS ----
def combined_loss(preds, targets):
    bce = bce_loss_fn(preds, targets)
    dice = dice_loss(preds, targets)
    focal = focal_loss(preds, targets)

    return 0.4 * bce + 0.3 * dice + 0.3 * focal


# ---- TRAIN ----
def train_one_epoch(model, loader, optimizer, device):
    model.train()
    total_loss = 0

    for imgs, masks in loader:
        imgs = imgs.float().to(device)
        masks = masks.to(device)

        preds = model(imgs)  # logits

        loss = combined_loss(preds, masks)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)


# ---- VALIDATE ----
def validate(model, loader, device, threshold=0.5):
    model.eval()
    total_loss = 0

    all_preds = []
    all_targets = []

    with torch.no_grad():
        for imgs, masks in loader:
            imgs = imgs.float().to(device)
            masks = masks.to(device)

            preds = model(imgs)  # logits

            loss = combined_loss(preds, masks)
            total_loss += loss.item()

            # ---- THRESHOLDING ----
            probs = torch.sigmoid(preds)
            preds_bin = (probs > threshold).float()

            all_preds.append(preds_bin.cpu())
            all_targets.append(masks.cpu())

    return total_loss / len(loader), all_preds, all_targets
```

```python
import matplotlib.pyplot as plt
import torch

EPOCHS = 40
PATIENCE = 5   # 🔥 stop if no improvement for 5 epochs

train_losses = []
val_losses = []

best_val_loss = float("inf")
counter = 0   # for early stopping

for epoch in range(EPOCHS):

    # ---- TRAIN ----
    train_loss = train_one_epoch(model, train_loader, optimizer, device)

    # ---- VALIDATE ----
    val_loss, preds, targets = validate(model, val_loader, device, threshold=0.3)

    train_losses.append(train_loss)
    val_losses.append(val_loss)

    # ---- SCHEDULER ----
    scheduler.step(val_loss)

    # ---- CHECK IMPROVEMENT ----
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        counter = 0   # reset patience
        torch.save(model.state_dict(), "best_lane_model.pth")
    else:
        counter += 1

    print(
        f"Epoch [{epoch+1}/{EPOCHS}] "
        f"Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} "
        f"| Patience: {counter}/{PATIENCE}"
    )

    # ---- EARLY STOPPING ----
    if counter >= PATIENCE:
        print("Early stopping triggered")
        break
```

```python
import cv2
import numpy as np

def postprocess_mask(pred, thresh=0.3):
    """
    pred: numpy array (H, W) with values in [0,1]
    """

    # ---- THRESHOLD ----
    mask = (pred > thresh).astype(np.uint8)

    # ---- SMALL KERNEL (lane-friendly) ----
    kernel = np.ones((3, 3), np.uint8)

    # ---- REMOVE NOISE ----
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    # ---- CONNECT BROKEN LANES ----
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    # ---- OPTIONAL: LIGHT SMOOTHING ----
    mask = cv2.medianBlur(mask * 255, 5)
    mask = (mask > 0).astype(np.uint8)

    return mask
```

```python
import torch
import numpy as np

def compute_metrics(pred, target):
    pred = pred.view(-1)
    target = target.view(-1)

    tp = (pred * target).sum().item()
    fp = (pred * (1 - target)).sum().item()
    fn = ((1 - pred) * target).sum().item()

    precision = tp / (tp + fp + 1e-8)
    recall    = tp / (tp + fn + 1e-8)
    dice      = (2 * tp) / (2 * tp + fp + fn + 1e-8)
    iou       = tp / (tp + fp + fn + 1e-8)

    return precision, recall, dice, iou


def evaluate_model(model, loader, device, thresholds=[0.2, 0.3, 0.4, 0.5]):
    model.eval()

    results = {}

    with torch.no_grad():
        for t in thresholds:
            total_p, total_r, total_d, total_i = 0, 0, 0, 0
            count = 0

            for imgs, masks in loader:
                imgs = imgs.float().to(device)
                masks = masks.to(device)

                preds = torch.sigmoid(model(imgs))
                preds = (preds > t).float()

                for p, m in zip(preds, masks):
                    precision, recall, dice, iou = compute_metrics(p, m)

                    total_p += precision
                    total_r += recall
                    total_d += dice
                    total_i += iou
                    count += 1

            results[t] = {
                "Precision": total_p / count,
                "Recall": total_r / count,
                "Dice": total_d / count,
                "IoU": total_i / count
            }

    return results
```

```python
results = evaluate_model(model, val_loader, device)

for t, metrics in results.items():
    print(f"\nThreshold = {t}")
    for k, v in metrics.items():
        print(f"{k}: {v:.4f}")
```

```python
best_t = max(results, key=lambda t: results[t]["Dice"])

print(f"\n🔥 Best Threshold: {best_t}")
print(results[best_t])
```

```python
import cv2
import numpy as np
import torch

IMG_H, IMG_W = 256, 512  # same as training

def preprocess_frame(frame):
    # ---- RESIZE ----
    frame = cv2.resize(frame, (IMG_W, IMG_H))

    # ---- BGR → RGB ----
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # ---- NORMALIZE ----
    frame = frame.astype(np.float32) / 255.0

    # ---- TO TENSOR ----
    tensor = torch.from_numpy(frame).permute(2, 0, 1)  # (H,W,C) → (C,H,W)

    # ---- ADD BATCH ----
    tensor = tensor.unsqueeze(0)

    return tensor
```

```python
def overlay_lane(frame, mask, alpha=0.6):
    mask = cv2.resize(mask, (frame.shape[1], frame.shape[0]))
    lane = np.zeros_like(frame)
    lane[:, :, 1] = (mask * 255).astype(np.uint8)  # green lanes
    return cv2.addWeighted(frame, 1.0, lane, alpha, 0)
```

```python
import cv2
import numpy as np

def overlay_lane(frame, mask, alpha=0.4):
    # ---- RESIZE MASK ----
    mask = cv2.resize(mask, (frame.shape[1], frame.shape[0]))

    # ensure binary
    mask = (mask > 0).astype(np.uint8)

    # ---- CREATE GREEN LANE ----
    lane = np.zeros_like(frame, dtype=np.uint8)
    lane[:, :, 1] = mask * 255  # green channel

    # ---- BLEND ----
    overlay = cv2.addWeighted(frame, 1.0, lane, alpha, 0)

    return overlay
```

```python
import cv2
import torch

video_path = "/content/input_lane_video.mp4"
out_path   = "/content/output_lane_video.mp4"

cap = cv2.VideoCapture(video_path)

# ---- SAFE FPS ----
fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0:
    fps = 25

width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(out_path, fourcc, fps, (width, height))

# ---- IMPORTANT ----
model.eval()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # ---- PREPROCESS ----
    inp = preprocess_frame(frame).to(device)

    with torch.no_grad():
        pred = torch.sigmoid(model(inp))[0, 0].cpu().numpy()

    # ---- CORRECT POSTPROCESS ----
    clean = postprocess_mask(pred, thresh=0.4)   # 🔥 use best threshold

    # ---- OVERLAY ----
    frame_out = overlay_lane(frame, clean)

    out.write(frame_out)

cap.release()
out.release()

print("✅ Video saved at:", out_path)
```
